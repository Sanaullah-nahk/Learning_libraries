```python
df1 = sns.load_dataset('car_crashes')
print(df1.shape)
df1.head()
```

# Distribution plot

```python
sns.displot(df1['not_distracted'], kde=True, bins=25 );
```

# joint plot

```python
sns.jointplot(x='speeding', y='alcohol', data=df1, kind='hex')
```

# KDE plot

```python
sns.kdeplot(df1['alcohol']);
```

# pairplots

```python
sns.pairplot(df1);
```
