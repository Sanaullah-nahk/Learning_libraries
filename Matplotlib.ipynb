```python
# https://www.youtube.com/watch?v=wB9C0Mz9gSo

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

```

# function plots

```python
x_1 = np.linspace(0,5,10)
y_1 = x_1**2

plt.plot(x_1, y_1)
plt.title('Days Squared Chart')

plt.xlabel('Days')
plt.ylabel('Days squared')
```

# Multiple plots

```python
fig1 = plt.figure(figsize=(10,6))
plt.tight_layout()
fig1.subplots_adjust( wspace=.9) #width space
kwargs_main  = { 'fontsize':16} 
fig1.suptitle('This is a somewhat long figure title', **kwargs_main )


##### Apple Plot
ax1 = plt.subplot(1,2,1)
plt.plot(x_1, y_1, 'r')
ax1.set_title('Apple')
plt.grid(True, color = '0.5')   #add grid

##### Banana Plot
ax2 = plt.subplot(1,2,2)
plt.plot(x_1, y_1, 'b')
plt.grid(True, color = '0.8', dashes=(5,5,0,0))   #add grid
ax2.set_facecolor('linen')
ax2.set_title('Banana');
```

# Using Figure Objects

```python
fig1 = plt.figure(figsize=(5,4), dpi=100)
axes_1 = fig1.add_axes([0.1, 0.1, 0.9,0.9])   # ***

axes_1.set_xlabel('Days')
axes_1.set_ylabel('Days Squared')
axes_1.set_title('Days Squared Chart')

axes_1.plot(x_1, y_1, label = 'x/x²')
axes_1.plot(y_1, x_1, label = 'x²/x')

axes_1.legend(loc=0)

# *** Adds axes with a   left,   bottom,   width,   height      that ranges from 0 to 1
#     which is the percent of the canvas you want to use

###################################################
# Adds axes with a left, bottom, width and height 
axes_2 = fig1.add_axes([0.38, 0.39, 0.44,0.5])
axes_2.set_xlabel('Days')
axes_2.set_ylabel('Days Squared')
axes_2.set_title('Days Squared Chart')
axes_2.plot(x_1, y_1, 'r', label = 'x/x²')
axes_2.legend(loc=0)




```

# sub plots


```python
fig2, ax2 = plt.subplots(figsize=(10,6), nrows=2, ncols=3 )
plt.tight_layout()
fig2.subplots_adjust(hspace=.4, wspace=.26)

kwargs_main  = {'x':0.5, 'y':1.06, 'fontsize':16} 
fig2.suptitle('This is a somewhat long figure title', **kwargs_main )

ax2[0,0].set_title('plot [0,0]' ,fontsize=14);
ax2[0,0].set_xlabel('Days');
ax2[0,0].set_ylabel('Days squared');

ax2[0,0].set_xticks(range(0,26,5))
ax2[0,0].set_yticks(range(0,26,2))

ax2[0,0].plot(x_1,y_1, label='x / x²',marker='.');
ax2[0,0].plot(y_1,x_1, label='x² / x', marker='.');
ax2[0,0].legend(loc=0);

ax2[0,0].text(6,15,'Multi plot');

########################################################
ax2[0,1].set_title('plot [0,1]', fontsize=14);
ax2[0,1].set_xlabel('Days');
ax2[0,1].set_ylabel('Days squared');

ax2[0,1].scatter(x_1,y_1, label='x / x²');
ax2[0,1].text(1,15,'scatter Plot');
ax2[0,1].legend(loc=(0.05,0.85));

ax2[0,1].grid(True, color = '0.8', dashes=(5,5,0,0))
ax2[0,1].set_facecolor('linen')

########################################################
ax2[0,2].set_title('plot [0,2]', fontsize=14);
ax2[0,2].set_xlabel('Days');
ax2[0,2].set_ylabel('Days squared');

ax2[0,2].bar(x_1,y_1, width= 0.4)
ax2[0,2].plot(x_1,y_1, label='x / x²',marker='.', color='orange');
ax2[0,2].text(0,15,'Bar and line Plot');
ax2[0,2].legend(loc=0);

########################################################
gaussian_numbers = np.random.normal(size=10000)
ax2[1,0].set_title('plot [1,0]', fontsize=14);
ax2[1,0].set_xlabel('values');
ax2[1,0].set_ylabel('frequency');

ax2[1,0].hist(gaussian_numbers, bins=100, label = 'gaussian');
ax2[1,0].text(1.5,250,'Histogram');
ax2[1,0].text(-4,250,'density=False')
ax2[1,0].legend(loc=2);

########################################################
gaussian_numbers = np.random.randn(500)
ax2[1,1].set_title('plot [1,1]', fontsize=14);
ax2[1,1].set_xlabel('values');
ax2[1,1].set_ylabel('Density');

ax2[1,1].hist(gaussian_numbers, bins=30, density=True, stacked=True, edgecolor="#6A9662", color="#DDFFDD", label = 'normal');
ax2[1,1].legend(loc=0);

##########################################################
ax2[1,2].set_title('plot [1,2]', fontsize=14);
ax2[1,2].set_xlabel('Days');
ax2[1,2].set_ylabel('Days squared');

ax2[1,2].step(x_1,y_1, label='x / x²');
ax2[1,2].text(1,15,'Step Plot');
ax2[1,2].legend(loc=(0.05,0.85));



```

### Appearance Options

```python
fig3 = plt.figure(figsize=(6,4))
ax3 = fig3.add_axes([0,0,1,1]) #Take Full axes
ax3.plot(x_1, y_1,
         color='navy', alpha=.9, linewidth=2, linestyle='-.',                                          #line style
         marker = 'o', markersize=7, markerfacecolor = 'y', markeredgecolor= 'r', markeredgewidth='2') # Marker Settings


ax3.set_xlim([0,3])    #to zoom in on X
ax3.set_ylim([0,8])    # to zoom in on Y

ax3.grid(True, color = '0.8', dashes=(5,5,0,0))   #add grid
ax3.set_facecolor('linen')                        #change backgroud
```
